```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("sheets/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US)')
plt.scatter(df.area, df.price , color = 'red' , marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']] , df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US)')
plt.scatter(df.area, df.price , color = 'red' , marker = '+')
#to predict the line
plt.plot(df.area , reg.predict(df[['area']]) , color = 'blue')
```
