```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("sheets/homeprices3.csv")
df.head()
```

```python
dummies = pd.get_dummies(df.town)
print(dummies)
```

```python
merged = pd.concat([df , dummies] , axis = 'columns')
merged
```

```python
# The Key advisory is the we have to drop the one of the dummy variable column
final = merged.drop(['town' , 'west windsor'] , axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop(['price'] , axis = 'columns')
X
```

```python
Y = final.price
Y
```

```python
model.fit(X , Y)
```

```python
model.predict([[2800 ,0,1]])
```

```python
model.score(X , Y)
```
